### Bollinger Bands

```python
# Bollinger Bands calculation method
def calculate_bollinger_bands(df, window):
    df.loc[:, 'MA'] = df.loc[:, ('Price')].rolling(window=window).mean()
    df['STD'] = df['Price'].rolling(window=window).std()
    #df['Upper_Band'] = df['MA'] + (df['STD'] * 2)
    #df['Lower_Band'] = df['MA'] - (df['STD'] * 2)

    df['Upper_Band'] = df['MA'] + (df['STD'])   # The above gives a big range, so there is no buying/selling trigger generated
    df['Lower_Band'] = df['MA'] - (df['STD'])
    return df
```

```python
N_DAYS = 7  # The past X number of days that will be used to predict the next day data
```

```python
# Calculating Bollinger Bands for our data
window = 20
df1 = calculate_bollinger_bands(df1, window=window)
```

### Build the model

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

model = Sequential([
    LSTM(60, return_sequences=True, input_shape=(X_train.shape[1], 1)),
    Dropout(0.3),
    LSTM(120, return_sequences=False),
    Dropout(0.3),
    Dense(20, activation='relu'),
    Dense(1, activation='linear')
])
```

### Compile the model

```python
model.compile(optimizer='adam', loss='mse')#, metrics=['accuracy'])
model.summary()
```

### Train the model

```python
epochs = 10
#history = model.fit(X_train, y_train, epochs=epochs, validation_data=(X_test, y_test))#, batch_size=N_DAYS)
history = model.fit(X_train, y_train, epochs=epochs)
```

#### Visualize training results

```python
loss = history.history['loss']
epochs_range = range(epochs)

plt.figure(figsize=(12, 6))
#plt.subplot(1, 3, 2)
plt.plot(epochs_range, loss, label='Training Loss')
#plt.plot(epochs_range, val_loss, label='Validation Loss')
#plt.xticks([0, 1, 2])
plt.title('epoch --- Loss', color='red')
plt.xlabel('epoch', color='red')
plt.ylabel('Loss', color='red')
plt.legend(loc='upper right')
plt.show()
```

### Making Prediction using Test data

```python
predictions = model.predict(X_test)
predictions = scaler.inverse_transform(predictions)

# Inverse transform the actual prices as well for comparison
actual_prices = scaler.inverse_transform(y_test.reshape(-1, 1))
```

### Make recommendations

```python
# Make recommendations using Bollinger Bands
df1_test['Recommendation'] = np.where(df1_test['predicted_price'] >= df1_test['Upper_Band'], 'Sell',
                        np.where(df1_test['predicted_price'] <= df1_test['Lower_Band'], 'Buy', 'Hold'))
```

#### Calculating 'Returns on Investment'

```python
initial_capital = 100000
df1_test['stocks_bought'] = 0

#df1_test.loc[df1_test['Recommendation'] == 'Buy', 'Position'] = 1     # implemented the below alternate code to check availability of stock before selling
#df1_test.loc[df1_test['Recommendation'] == 'Sell', 'Position'] = -1

# Initialize a variable to track whether stocks were bought
stocks_held = False

# Iterate over the DataFrame rows to check availability of stocks before selling and set the value of stocks_bought based on the Recommendation
for i in range(len(df1_test)):
    if df1_test.loc[i, 'Recommendation'] == 'Buy':
        df1_test.loc[i, 'stocks_bought'] = 1
        stocks_held = True
    elif df1_test.loc[i, 'Recommendation'] == 'Sell' and stocks_held:
        df1_test.loc[i, 'stocks_bought'] = -1
        stocks_held = False

df1_test['Daily_Return'] = df1_test['Price'].pct_change()
df1_test['Strategy_Return'] = df1_test['Daily_Return'] * df1_test['stocks_bought'].shift(1)
df1_test['Capital'] = initial_capital * (1 + df1_test['Strategy_Return']).cumprod()
```
